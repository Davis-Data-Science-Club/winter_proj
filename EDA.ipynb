```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Tercero = pd.read_csv("df_tercero_all.csv")
```

```python
Tercero
```

```python
def calc_ratio (x, y):
    if y <= 0:
        y = None
        return None
    return (x / y)

# apply_CalCarbs
# Tercero
```

```python
apply_CalCarbs = Tercero.apply (
    lambda x: calc_ratio(x['Carbohydrates'], x['Calories']), axis=1)
Tercero['CalCarbs'] = apply_CalCarbs

apply_CalFat = Tercero.apply (
    lambda x: calc_ratio(x['Fat'], x['Calories']), axis=1)
Tercero['CalFat'] = apply_CalFat

apply_CalPro = Tercero.apply (
    lambda x: calc_ratio(x['Protein'], x['Calories']), axis=1)
Tercero['CalPro'] = apply_CalPro
```

```python
CalCarbsGraph = sns.lineplot(x=Tercero['Calories'], y=Tercero['Carbohydrates'], data = Tercero, hue = Tercero['Mealtime'])
CalCarbsGraph.set(xlim=(0,800) , ylim=(0,150))
plt.show()
```

```python
CalCarbsGraph = sns.lineplot(x=Tercero['Calories'], y=Tercero['Carbohydrates'], data = Tercero, hue = Tercero['Location'])
CalCarbsGraph.set(xlim=(0, 800) , ylim=(0,300))
plt.show()

# make upper fence function to set range limit for x and y
```
